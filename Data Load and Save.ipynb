# Data Save

```python
import pandas as pd
import os
os.chdir("/camin1/inyoung/DATA/data_csv")
```

```python
list_name = os.listdir()
os.chdir("/camin1/inyoung/")
```

```python
for i in list_name:
    name = i.split(".")[0]
    predict = pd.read_csv("./MTAD-GAT/result/" + str(name) + "_iter_0_testresult.csv")["Pred_Global"]
    data = pd.read_csv("./DATA/data_csv/" + str(name) + ".csv")
    result = pd.DataFrame({'Time':data.index[64:]+1})
    result = pd.concat([result, data.iloc[64:,:].reset_index(drop=True), predict],axis=1)
    result.to_csv("./Total Result/result_"+str(name)+".csv",index=False)
```
