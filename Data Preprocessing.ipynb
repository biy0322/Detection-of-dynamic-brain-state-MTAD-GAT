```python
import os
import warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd

import sranodec as anom


from tqdm import tqdm
from sklearn.preprocessing import MinMaxScaler

import seaborn as sns
import matplotlib.pyplot as plt
```

# Data Preprocessing

```python
os.chdir("/camin1/inyoung/DATA/data_csv")
data_list = os.listdir()
```

```python
def data_preprocessing(name):
    train_x = pd.read_csv(name, header=0)
    test_x = pd.read_csv(name, header=0)
    
    ############# Normalization #############
    scaler = MinMaxScaler()
    scaler.fit(train_x.values)
    train_x[train_x.columns] = scaler.transform(train_x.values)
    test_x[test_x.columns] = scaler.transform(test_x.values)
    
    ############# Cleaning SR #############
    sr = True
    if sr:
        for col in train_x.columns:
            ### amp_window_size, series_window_size, score_window_size
            spec = anom.Silency(24,24,100)
            score = spec.generate_anomaly_score(train_x[col].values)
            train_x[col].values[score > np.percentile(score, 99)] = np.NaN
            
    ############# Interpolate #############
    train_x = train_x.interpolate(method="linear")
    
    ############# Save Data #############
    save_path = "/camin1/inyoung/DATA/data_pre/"
    train_x.to_csv(save_path+"train_"+name, index=False)
    test_x.to_csv(save_path+"test_"+name, index=False)
```

```python
for name in data_list:
    print("################### Start " + name + "###################")
    data_preprocessing(name)
    print("################### Finish ###################")
```
